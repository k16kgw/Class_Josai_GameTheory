
# 第3回 演習・課題 解答確認ノートブック

このノートブックは，次の問題の解答を **GUI で入力し，その場で確認** するためのものです．

- 演習1：戦略と戦略プロフィール
- 演習2：混合戦略の期待利得
- 課題1：ペナルティキックの期待利得
- 演習3：クールノー寡占市場の利潤関数
- 課題2：具体的なクールノー寡占市場

## 使い方
各セルを上から順に実行してください．  
各問題の入力欄に答えを入れて，**判定** ボタンを押すと結果が表示されます．

```{note}
数値判定では，小数の丸め誤差を考慮して近い値も正解として扱います．
```


````python
import math
import re
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

TOL = 1e-6

def normalize_text(s):
    s = str(s).strip()
    s = s.replace(" ", "").replace("　", "")
    return s

def parse_profiles(text):
    # (A,C),(A,D) のような入力を想定
    s = normalize_text(text)
    pairs = re.findall(r'\(([^,()]+),([^,()]+)\)', s)
    return [(a, b) for a, b in pairs]

def approx_equal(a, b, tol=1e-6):
    return abs(a - b) <= tol

def parse_float(s):
    return float(str(s).strip())

def show_result(lines, ok=True):
    color = "green" if ok else "crimson"
    html = "<br>".join(lines)
    display(widgets.HTML(value=f"<div style='color:{color}; line-height:1.6'>{html}</div>"))

print("準備完了")
````

## 演習1　戦略と戦略プロフィール

````python

p1_strategies = widgets.Text(
    description='P1戦略',
    placeholder='例: A,B',
    layout=widgets.Layout(width='420px')
)
p2_strategies = widgets.Text(
    description='P2戦略',
    placeholder='例: C,D,E',
    layout=widgets.Layout(width='420px')
)
profiles = widgets.Textarea(
    description='プロフィール',
    placeholder='例: (A,C),(A,D),(A,E),(B,C),(B,D),(B,E)',
    layout=widgets.Layout(width='700px', height='90px')
)
count = widgets.BoundedIntText(
    description='総数',
    min=0, max=100, value=0,
    layout=widgets.Layout(width='220px')
)
btn1 = widgets.Button(description='演習1を判定', button_style='primary')
out1 = widgets.Output()

def check_ex1(_):
    with out1:
        clear_output()
        ok = True
        lines = []

        ans_p1 = [x for x in normalize_text(p1_strategies.value).split(",") if x]
        ans_p2 = [x for x in normalize_text(p2_strategies.value).split(",") if x]
        ans_profiles = set(parse_profiles(profiles.value))
        true_p1 = ["A", "B"]
        true_p2 = ["C", "D", "E"]
        true_profiles = {("A","C"),("A","D"),("A","E"),("B","C"),("B","D"),("B","E")}
        true_count = 6

        if ans_p1 == true_p1:
            lines.append("P1 の戦略：正解です．")
        else:
            ok = False
            lines.append(f"P1 の戦略：不正解です．正しくは {true_p1} です．")

        if ans_p2 == true_p2:
            lines.append("P2 の戦略：正解です．")
        else:
            ok = False
            lines.append(f"P2 の戦略：不正解です．正しくは {true_p2} です．")

        if ans_profiles == true_profiles:
            lines.append("戦略プロフィール：正解です．")
        else:
            ok = False
            lines.append(f"戦略プロフィール：不正解です．正しくは {sorted(true_profiles)} です．")

        if count.value == true_count:
            lines.append("総数：正解です．")
        else:
            ok = False
            lines.append(f"総数：不正解です．正しくは {true_count} です．")

        show_result(lines, ok)

btn1.on_click(check_ex1)

display(widgets.VBox([p1_strategies, p2_strategies, profiles, count, btn1, out1]))
````

## 演習2　混合戦略の期待利得


期待利得の正解は
$$
E[f_1]=4pq-2p-2q+1,\qquad E[f_2]=-4pq+2p+2q-1
$$
です．

ここでは，展開後の係数を入力して確認します．  
すなわち
$$
E[f_1]=a\,pq+b\,p+c\,q+d,\qquad
E[f_2]=e\,pq+f\,p+g\,q+h
$$
の係数 $a,b,c,d,e,f,g,h$ を答えてください．


````python

style = {'description_width': '40px'}
a = widgets.FloatText(description='a', style=style, layout=widgets.Layout(width='120px'))
b = widgets.FloatText(description='b', style=style, layout=widgets.Layout(width='120px'))
c = widgets.FloatText(description='c', style=style, layout=widgets.Layout(width='120px'))
d = widgets.FloatText(description='d', style=style, layout=widgets.Layout(width='120px'))

e = widgets.FloatText(description='e', style=style, layout=widgets.Layout(width='120px'))
f = widgets.FloatText(description='f', style=style, layout=widgets.Layout(width='120px'))
g = widgets.FloatText(description='g', style=style, layout=widgets.Layout(width='120px'))
h = widgets.FloatText(description='h', style=style, layout=widgets.Layout(width='120px'))

btn2 = widgets.Button(description='演習2を判定', button_style='primary')
out2 = widgets.Output()

def check_ex2(_):
    with out2:
        clear_output()
        true_vals = {'a':4, 'b':-2, 'c':-2, 'd':1, 'e':-4, 'f':2, 'g':2, 'h':-1}
        user_vals = {'a':a.value, 'b':b.value, 'c':c.value, 'd':d.value,
                     'e':e.value, 'f':f.value, 'g':g.value, 'h':h.value}
        ok = True
        lines = []
        for key in true_vals:
            if approx_equal(user_vals[key], true_vals[key]):
                lines.append(f"{key}：正解")
            else:
                ok = False
                lines.append(f"{key}：不正解（正しくは {true_vals[key]}）")
        show_result(lines, ok)

btn2.on_click(check_ex2)

display(widgets.HTML("<b>E[f_1]=a pq + b p + c q + d</b>"))
display(widgets.HBox([a,b,c,d]))
display(widgets.HTML("<b>E[f_2]=e pq + f p + g q + h</b>"))
display(widgets.HBox([e,f,g,h]))
display(widgets.VBox([btn2, out2]))
````

## 課題1　ペナルティキックの期待利得

````python

keeper = widgets.FloatText(description='キーパー', layout=widgets.Layout(width='180px'))
kicker = widgets.FloatText(description='キッカー', layout=widgets.Layout(width='180px'))
btn3 = widgets.Button(description='課題1を判定', button_style='primary')
out3 = widgets.Output()

def check_hw1(_):
    with out3:
        clear_output()
        true_keeper = -0.08
        true_kicker = 0.08
        ok1 = approx_equal(keeper.value, true_keeper, 1e-4)
        ok2 = approx_equal(kicker.value, true_kicker, 1e-4)
        ok = ok1 and ok2
        lines = []
        lines.append(f"キーパー：{'正解' if ok1 else f'不正解（正しくは {true_keeper}）'}")
        lines.append(f"キッカー：{'正解' if ok2 else f'不正解（正しくは {true_kicker}）'}")
        show_result(lines, ok)

btn3.on_click(check_hw1)
display(widgets.VBox([keeper, kicker, btn3, out3]))
````

## 演習3　クールノー寡占市場の利潤関数


正解は
$$
f_1(q_1,q_2)=\bigl(a-b(q_1+q_2)\bigr)q_1-cq_1,\qquad
f_2(q_1,q_2)=\bigl(a-b(q_1+q_2)\bigr)q_2-cq_2
$$
です．

ここでは展開後の係数で確認します．
$$
f_1 = \alpha q_1 + \beta q_1^2 + \gamma q_1 q_2,\qquad
f_2 = \delta q_2 + \varepsilon q_2^2 + \zeta q_1 q_2
$$
ただし，係数の中に $a,b,c$ が入ります．


````python

alpha = widgets.Text(description='α', placeholder='例: a-c', layout=widgets.Layout(width='180px'))
beta  = widgets.Text(description='β', placeholder='例: -b', layout=widgets.Layout(width='180px'))
gamma = widgets.Text(description='γ', placeholder='例: -b', layout=widgets.Layout(width='180px'))
delta = widgets.Text(description='δ', placeholder='例: a-c', layout=widgets.Layout(width='180px'))
epsi  = widgets.Text(description='ε', placeholder='例: -b', layout=widgets.Layout(width='180px'))
zeta  = widgets.Text(description='ζ', placeholder='例: -b', layout=widgets.Layout(width='180px'))

btn4 = widgets.Button(description='演習3を判定', button_style='primary')
out4 = widgets.Output()

def canonical(expr):
    s = normalize_text(expr)
    # よくある同値表現を雑に統一
    s = s.replace("q_1","q1").replace("q_2","q2")
    s = s.replace("c-a","-(a-c)")
    s = s.replace("-1*b","-b")
    return s

def eq_text(ans, target_options):
    s = canonical(ans)
    return s in {canonical(t) for t in target_options}

def check_ex3(_):
    with out4:
        clear_output()
        checks = [
            (alpha.value, ["a-c"]),
            (beta.value, ["-b"]),
            (gamma.value, ["-b"]),
            (delta.value, ["a-c"]),
            (epsi.value, ["-b"]),
            (zeta.value, ["-b"]),
        ]
        labels = ["α","β","γ","δ","ε","ζ"]
        ok = True
        lines = []
        truths = ["a-c","-b","-b","a-c","-b","-b"]
        for lab, (ans, targets), tru in zip(labels, checks, truths):
            if eq_text(ans, targets):
                lines.append(f"{lab}：正解")
            else:
                ok = False
                lines.append(f"{lab}：不正解（正しくは {tru}）")
        show_result(lines, ok)

btn4.on_click(check_ex3)
display(widgets.VBox([
    widgets.HTML("<b>f1 = α q1 + β q1^2 + γ q1 q2</b>"),
    widgets.HBox([alpha, beta, gamma]),
    widgets.HTML("<b>f2 = δ q2 + ε q2^2 + ζ q1 q2</b>"),
    widgets.HBox([delta, epsi, zeta]),
    btn4, out4
]))
````

## 課題2　具体的なクールノー寡占市場

````python

f1_expr = widgets.Text(
    description='f1',
    placeholder='例: (10-q1-q2)q1',
    layout=widgets.Layout(width='340px')
)
f2_expr = widgets.Text(
    description='f2',
    placeholder='例: (10-q1-q2)q2',
    layout=widgets.Layout(width='340px')
)
quad_a = widgets.FloatText(description='q1係数', layout=widgets.Layout(width='180px'))
quad_b = widgets.FloatText(description='q1^2係数', layout=widgets.Layout(width='180px'))
quad_c = widgets.FloatText(description='q1q2係数', layout=widgets.Layout(width='180px'))
effect = widgets.Dropdown(
    description='影響',
    options=[
        ('選択してください', ''),
        ('相手企業の生産量が増えると，自社の利潤は減少する', 'decrease'),
        ('相手企業の生産量が増えると，自社の利潤は増加する', 'increase'),
        ('影響しない', 'none')
    ],
    layout=widgets.Layout(width='520px')
)
reason = widgets.Text(
    description='理由',
    placeholder='例: 総供給量が増えて価格が下がるため',
    layout=widgets.Layout(width='700px')
)

btn5 = widgets.Button(description='課題2を判定', button_style='primary')
out5 = widgets.Output()

def normalize_expr(s):
    s = normalize_text(s)
    s = s.replace("*","")
    return s

def check_hw2(_):
    with out5:
        clear_output()
        ok = True
        lines = []

        target_f1 = {"(10-q1-q2)q1", "q1(10-q1-q2)"}
        target_f2 = {"(10-q1-q2)q2", "q2(10-q1-q2)"}
        ok_f1 = normalize_expr(f1_expr.value) in {normalize_expr(x) for x in target_f1}
        ok_f2 = normalize_expr(f2_expr.value) in {normalize_expr(x) for x in target_f2}
        ok_q1 = approx_equal(quad_a.value, 10)
        ok_q1sq = approx_equal(quad_b.value, -1)
        ok_cross = approx_equal(quad_c.value, -1)
        ok_eff = (effect.value == 'decrease')
        reason_ok = ('価格' in reason.value) or ('供給' in reason.value) or ('下が' in reason.value)

        checks = [
            ("f1", ok_f1, "(10-q1-q2)q1"),
            ("f2", ok_f2, "(10-q1-q2)q2"),
            ("q1係数", ok_q1, "10"),
            ("q1^2係数", ok_q1sq, "-1"),
            ("q1q2係数", ok_cross, "-1"),
            ("影響", ok_eff, "相手企業の生産量が増えると，自社の利潤は減少する"),
            ("理由", reason_ok, "総供給量が増えて価格が下がるため，で説明できるとよい"),
        ]
        for label, judge, truth in checks:
            if judge:
                lines.append(f"{label}：正解")
            else:
                ok = False
                lines.append(f"{label}：不正解（目安：{truth}）")
        show_result(lines, ok)

btn5.on_click(check_hw2)

display(widgets.VBox([
    widgets.HTML("<b>1. 利潤関数</b>"),
    f1_expr, f2_expr,
    widgets.HTML("<b>2. f1 を q1 の二次式で見たときの係数</b>"),
    widgets.HBox([quad_a, quad_b, quad_c]),
    widgets.HTML("<b>3. 相手企業の生産量が増えるとどうなるか</b>"),
    effect, reason,
    btn5, out5
]))
````
