```python
actions = ["C", "D"]

U1 = [
    [3, 0],
    [4, 1],
]

U2 = [
    [3, 4],
    [0, 1],
]

def find_pure_nash(U1, U2, actions_p1, actions_p2):
    m = len(actions_p1)
    n = len(actions_p2)
    equilibria = []

    for i in range(m):
        for j in range(n):
            p1_best = U1[i][j] == max(U1[r][j] for r in range(m))
            p2_best = U2[i][j] == max(U2[i][c] for c in range(n))

            if p1_best and p2_best:
                equilibria.append((actions_p1[i], actions_p2[j], U1[i][j], U2[i][j]))

    return equilibria

equilibria = find_pure_nash(U1, U2, actions, actions)
print(equilibria)
```

```python
import numpy as np
import pandas as pd

def find_pure_nash(U1, U2, actions_p1, actions_p2):
    """
    2人有限ゲームの純戦略ナッシュ均衡を求める関数

    U1: プレイヤー1の利得行列
    U2: プレイヤー2の利得行列
    actions_p1: プレイヤー1の戦略名のリスト
    actions_p2: プレイヤー2の戦略名のリスト
    """
    U1 = np.array(U1)
    U2 = np.array(U2)

    equilibria = []

    for i in range(len(actions_p1)):
        for j in range(len(actions_p2)):
            p1_best = U1[i, j] == np.max(U1[:, j])
            p2_best = U2[i, j] == np.max(U2[i, :])

            if p1_best and p2_best:
                equilibria.append({
                    "P1_strategy": actions_p1[i],
                    "P2_strategy": actions_p2[j],
                    "P1_payoff": U1[i, j],
                    "P2_payoff": U2[i, j]
                })

    return equilibria
```

```python
equilibria = find_pure_nash(U1, U2, actions, actions)
equilibria
```

```python
def pareto_dominates(payoff_x, payoff_y):
    """
    payoff_x が payoff_y をパレート優位するか判定する．
    payoff_x, payoff_y は利得ベクトルを表すタプルまたはリスト．
    """
    all_weakly_better = all(x >= y for x, y in zip(payoff_x, payoff_y))
    at_least_one_strictly_better = any(x > y for x, y in zip(payoff_x, payoff_y))

    return all_weakly_better and at_least_one_strictly_better


def find_pareto_optimal_profiles(U1, U2, actions_p1, actions_p2):
    """
    2人有限ゲームにおけるパレート最適な戦略プロフィールを求める．
    """
    U1 = np.array(U1)
    U2 = np.array(U2)

    profiles = []

    for i, a1 in enumerate(actions_p1):
        for j, a2 in enumerate(actions_p2):
            profiles.append({
                "profile": (a1, a2),
                "payoff": (U1[i, j], U2[i, j])
            })

    pareto_optimal = []

    for x in profiles:
        dominated = False

        for y in profiles:
            if pareto_dominates(y["payoff"], x["payoff"]):
                dominated = True
                break

        if not dominated:
            pareto_optimal.append(x)

    return pareto_optimal
```

```python
pareto_profiles = find_pareto_optimal_profiles(U1, U2, actions, actions)
pareto_profiles
```

```python
def price(Q):
    return 12 - Q

def profit_1(q1, q2):
    Q = q1 + q2
    return (price(Q) - 2) * q1

def profit_2(q1, q2):
    Q = q1 + q2
    return (price(Q) - 2) * q2

def total_profit(q1, q2):
    return profit_1(q1, q2) + profit_2(q1, q2)
```

```python
q_star = 10 / 3

print("q1* =", q_star)
print("q2* =", q_star)
print("Q* =", 2 * q_star)
print("p(Q*) =", price(2 * q_star))
print("f1(q*) =", profit_1(q_star, q_star))
print("f2(q*) =", profit_2(q_star, q_star))
print("total profit =", total_profit(q_star, q_star))
```

```python
q_max = 5 / 2

print("q1_max =", q_max)
print("q2_max =", q_max)
print("Q_max =", 2 * q_max)
print("p(Q_max) =", price(2 * q_max))
print("f1(q_max) =", profit_1(q_max, q_max))
print("f2(q_max) =", profit_2(q_max, q_max))
print("total profit =", total_profit(q_max, q_max))
```
