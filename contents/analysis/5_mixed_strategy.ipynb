```python
def expected_payoffs(U1, U2, p, q):
    """
    p: player 1 chooses row 0 with probability p
    q: player 2 chooses column 0 with probability q
    """
    probs = [
        [p * q, p * (1 - q)],
        [(1 - p) * q, (1 - p) * (1 - q)],
    ]

    e1 = 0
    e2 = 0
    for i in range(2):
        for j in range(2):
            e1 += probs[i][j] * U1[i][j]
            e2 += probs[i][j] * U2[i][j]

    return e1, e2

U1 = [
    [1, -1],
    [-1, 1],
]

U2 = [
    [-1, 1],
    [1, -1],
]

print(expected_payoffs(U1, U2, p=0.5, q=0.5))
print(expected_payoffs(U1, U2, p=0.8, q=0.5))
```
