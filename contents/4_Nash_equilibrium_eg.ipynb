```python
actions_p1 = ["U", "D"]
actions_p2 = ["L", "R"]

U1 = [
    [4, 1],
    [2, 3],
]

U2 = [
    [3, 2],
    [1, 4],
]

def best_responses_p1(U1, actions_p1, actions_p2):
    responses = {}
    m = len(actions_p1)
    n = len(actions_p2)

    for j in range(n):
        column_values = [U1[i][j] for i in range(m)]
        max_value = max(column_values)
        responses[actions_p2[j]] = [
            actions_p1[i] for i in range(m) if U1[i][j] == max_value
        ]

    return responses

def best_responses_p2(U2, actions_p1, actions_p2):
    responses = {}
    m = len(actions_p1)

    for i in range(m):
        row_values = U2[i]
        max_value = max(row_values)
        responses[actions_p1[i]] = [
            actions_p2[j] for j in range(len(actions_p2)) if U2[i][j] == max_value
        ]

    return responses

def find_pure_nash(U1, U2, actions_p1, actions_p2):
    m = len(actions_p1)
    n = len(actions_p2)
    equilibria = []

    for i in range(m):
        for j in range(n):
            p1_best = U1[i][j] == max(U1[r][j] for r in range(m))
            p2_best = U2[i][j] == max(U2[i][c] for c in range(n))

            if p1_best and p2_best:
                equilibria.append((actions_p1[i], actions_p2[j], U1[i][j], U2[i][j]))

    return equilibria

br1 = best_responses_p1(U1, actions_p1, actions_p2)
br2 = best_responses_p2(U2, actions_p1, actions_p2)

print("プレイヤー1の最適応答")
print(br1)

print("プレイヤー2の最適応答")
print(br2)

equilibria = find_pure_nash(U1, U2, actions_p1, actions_p2)

print("ナッシュ均衡")
for eq in equilibria:
    print(eq)
```
